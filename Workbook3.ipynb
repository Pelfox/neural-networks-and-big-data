### 1.2.1 Пример

**Задача:**

Напишите функцию `sum_range(start, end)`, которая суммирует все целые
числа от значения «start» до величины «end» включительно. 
Если пользователь задаст первое число большее чем второе, просто 
поменяйте их местами.

```python
def sum_range(start: int, end: int) -> int:
    if start > end:
        end, start = start, end
    return sum(range(start, end + 1))

# Тесты
print(sum_range(2, 12))
print(sum_range(-4, 4))
print(sum_range(3, 2))
```

### 1.2.2 Пример

Напишите рекурсивную функцию вычисления факториала на языке 
Python.

```python
def factorial(number: int) -> int:
    if number == 0:
        return 1
    return number * factorial(number - 1)

# Тесты
print(factorial(5))
```

### 1.2.3 Пример

**Задача:**

Напишите функции в Python, которая вычисляет Евклидово расстояние 
между двумя массивами NumPy.

```python
import numpy as np

def euclidian_distance(v1: np.ndarray, v2: np.ndarray) -> float:
    return sum((x - y) ** 2 for x, y in zip(v1, v2)) ** 0.5

# Тесты
x = np.array([0, 0, 0])
y = np.array([3, 3, 3])
print(euclidian_distance(x, y))
```

### 1.2.4 Пример

**Задача:**

Напишите 4 функции в Python, которые рассчитывают квадрат Евклидова 
расстояния, взвешенное евклидово расстояние, Хеммингово расстояние и 
расстояние Чебышева между двумя массивами NumPy.

```python
def sqr_euclidian_distance(v1: np.ndarray, v2: np.ndarray) -> int:
    return sum((x - y) ** 2 for x, y in zip(v1, v2))

def weighted_euclidian_distance(v1: np.ndarray, v2: np.ndarray, weights: np.ndarray) -> float:
    return sum((x - y) ** 2 * s for x, y, s in zip(v1, v2, weights)) ** 0.5

def manhattan_distance(v1: np.ndarray, v2: np.ndarray) -> int:
    return sum(abs(x - y) for x, y in zip(v1, v2))

def chebyshev_distance(v1: np.ndarray, v2: np.ndarray) -> int:
    return max(abs(x - y) for x, y in zip(v1, v2))

# Тесты
x = np.array([0, 0, 0])
y = np.array([3, 3, 3])
weights = np.array([0, 0, 1])

print(sqr_euclidian_distance(x, y))
print(weighted_euclidian_distance(x, y, weights))
print(manhattan_distance(x, y))
print(chebyshev_distance(x, y))
```

Задайте 4 точки в трехмерном пространстве, рассчитайте между ними 
расстояния по описанным в примере выше метрикам. Отобразите точки 
в трехмерном пространстве.

```python
point_1 = np.array([2.3, -1.1, 0.7])
point_2 = np.array([-0.5, 3.2, 1.8])
point_3 = np.array([1.4, 0.9, -2.6])
point_4 = np.array([-3.1, -0.8, 2.4])

points = [point_1, point_2, point_3, point_4]
for outer_idx in range(0, len(points)):
    start_point = points[outer_idx]
    for inner_idx in range(0, len(points)):
        if inner_idx == outer_idx:
            continue
        end_point = points[inner_idx]
        print(f"Euclidian distance between {start_point} and {end_point} is:", np.linalg.norm(start_point - end_point))
        print(f"Squared Euclidian distance between {start_point} and {end_point} is:", np.linalg.norm(start_point - end_point) ** 2)
        print(f"Chebyshev distance between {start_point} and {end_point} is:", np.linalg.norm(start_point - end_point, ord=np.inf))
        print(f"Manhattan distance between {start_point} and {end_point} is:", np.linalg.norm(start_point - end_point, ord=1))
        print("===")

points_figure = plt.figure()
ax = points_figure.add_subplot(111, projection="3d")

for point in points:
    ax.scatter(*point)

plt.show()
```

### 2.2.1 Пример

**Задача:**

В примере показано создание 2d-массива со значениями x и y. Список
target содержит возможные выходные классы (часто называемые 
метками). Далее происходит обучение классификатора k-ближайших
соседей по исходным данным. Далее производится прогноз 
принадлежности к классам для двух точек данных.

```python
from sklearn.neighbors import KNeighborsClassifier

x = np.array([[-1, -1], [-2, -1], [-3, -2],
              [1, 1], [2, 1], [3, 2]])
target = [0, 0, 0, 1, 1, 1]

K = 3
model = KNeighborsClassifier(n_neighbors=K)
model.fit(x, target)
print("Model:", model)

print("(-2, -2) is class", model.predict([[-2, 2]]))
print("(1, 3) is class", model.predict([[1, 3]]))
```

### 2.2.4 Пример

**Задача:**

Из графиков видно, что в первом случае классы визуально хорошо 
отделимы друг от друга, хотя два класса имеют небольшое пересечение.
Во втором случае разделить два класса между собой уже намного труднее. 

Далее разделим датасет на обучающую и тестовую выборки в 
соотношении 80:20. Обучающая выборка (training sample) — выборка, по
которой производится настройка (оптимизация параметров) модели 
зависимости. Тестовая (или контрольная) выборка (test sample) — 
выборка, по которой оценивается качество построенной модели.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(
    iris.iloc[:, :-1],
    iris.iloc[:, -1],
    test_size=0.2,
)
display(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
display(x_train.head(), y_train.head())

model = KNeighborsClassifier(n_neighbors=3)
model.fit(x_train, y_train)

y_prediction = model.predict(x_test)
display(y_prediction)

plt.figure(figsize=(10, 7))
sns.scatterplot(
    x="petal_width",
    y="petal_length",
    data=iris,
    hue="species",
    s=70
)
plt.xlabel("Длина лепестка (см)")
plt.ylabel("Ширина лепестка (см)")
plt.legend(loc=2)
plt.grid()

for i in range(len(y_test)):
    if np.array(y_test)[i] != y_prediction[i]:
        plt.scatter(x_test.iloc[i, 3], x_test.iloc[i, 2], color="red", s=150)

print("Accuracy:", accuracy_score(y_test, y_prediction))
```

### 2.3.1. Задача

**Задача:**

Для предыдущего примера поэкспериментируйте с параметрами 
классификатора: 
1. Установите другое количество ближайших соседей (k = 1, 5, 10). 
2. Установите размер тестовой выборки 15% от всего датасета. 
3. Постройте графики и оцените качество моделей, проанализируйте 
результаты.

```python
results = {}

for k in [1, 5, 10]:
    x_train, x_test, y_train, y_test = train_test_split(
        iris.iloc[:, :-1],
        iris.iloc[:, -1],
        test_size=0.15,
    )

    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train, y_train)

    y_prediction = model.predict(x_test)
    results[k] = accuracy_score(y_test, y_prediction)

    plt.figure(figsize=(10, 7))
    sns.scatterplot(
        x="petal_width",
        y="petal_length",
        data=iris,
        hue="species",
        s=70
    )
    plt.title(f"K = {k}, accuracy = {results[k]}")
    plt.xlabel("Длина лепестка (см)")
    plt.ylabel("Ширина лепестка (см)")
    plt.legend(loc=2)
    plt.grid()

    for i in range(len(y_test)):
        if np.array(y_test)[i] != y_prediction[i]:
            plt.scatter(x_test.iloc[i, 3], x_test.iloc[i, 2], color="red", s=150)

for (k, accuracy) in results.items():
    print(f"{k}: {accuracy}")
```
