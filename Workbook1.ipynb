```python
# Задание 4.3.1
import random
import statistics
from matplotlib import pyplot as plt

COUNT = 100
numbers = [random.random() for _ in range(COUNT)]

avg = sum(numbers) / COUNT
median = statistics.median(numbers)

print('Среднее значение', avg)
print('Медианное значение', median)

plt.figure(figsize=(10, 7))
plt.scatter(range(COUNT), numbers)
plt.show()
```

```python
# Задание 4.3.2
from matplotlib import pyplot as plt

def f(x: int):
    return (math.sqrt(1 + math.pow(math.e, math.sqrt(x)) + math.cos(math.pow(x, 2))) / abs(1 - math.pow(math.sin(x), 3))) + math.log(abs(2 * x))

numbers = [f(x) for x in range(1, 11)]
first_half = numbers[:len(numbers) // 2]

plt.figure(figsize=(10, 7))
plt.plot(range(len(numbers)), numbers, label='Основной массив (линейный)')
plt.scatter(range(len(first_half)), first_half, color='red', s=50, label='Первая половина (точечный)')
plt.legend()
plt.show()
```

```python
# Задача 4.3.3
import math
from matplotlib import pyplot as plt
from scipy.integrate import simpson
from numpy import trapz

def f(x: int):
    return abs(math.cos(x * math.pow(math.e, math.cos(x) + math.log(x + 1))))

x_values = [x for x in range(0, 11)]
y_values = [f(x) for x in x_values]

area_trapz = trapz(y_values, x_values)
area_simpson = simpson(y_values, x_values)

plt.figure(figsize=(10, 7))
plt.plot(x_values, y_values)
plt.fill_between(x_values, y_values, label=f'Площадь = {area_trapz:.2f}')
plt.legend()
plt.show()

print(f"Площадь под кривой (метод трапеций): {area_trapz:.4f}")
print(f"Площадь под кривой (метод Симпсона): {area_simpson:.4f}")
```

```python
# Задача 4.3.4
import yfinance
from matplotlib import pyplot as plt

companies = {
    'AAPL': 'Apple',
    'MSFT': 'Microsoft',
    'GOOGL': 'Google',
}

data = {}
for company, name in companies.items():
    data[name] = yfinance.download(company, start='2021-01-01', end='2021-12-31')['Close']

plt.figure(figsize=(10, 7))

colors = ['blue', 'green', 'red']
for i, (company, prices) in enumerate(data.items()):
    plt.plot(prices.index, prices.values, label=company, color=colors[i])

plt.title('График изменения цен компаний')
plt.xlabel('Дата')
plt.ylabel('Цена закрытия (USD)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Задание 4.3.5
import math
from typing import Tuple

def request_numbers() -> Tuple[float, float]:
    return float(input('Введите первое число: ')), float(input('Введите второе число: '))

functions = {
    1: lambda x, y: math.pow(math.e, (x + y)),
    2: lambda x, y: math.sin(x + y),
    3: lambda x, y: math.cos(x + y),
    4: lambda x, y: x ** y,
}

action_text = 'Добро пожаловать в калькулятор.\n' \
    'Выберите, пожалуйста, операцию для продолжения: +, -, * или /\n' \
    'Либо, выберите функцию для расчёта:\n' \
    ' 1. e^{x+y}\n' \
    ' 2. sin(x + y)\n' \
    ' 3. cos(x + y)\n' \
    ' 4. x^y\n' \
    'Ваш выбор: '
action = input(action_text)

if action.isnumeric():
    action = int(action)
    if action < 0 or action > len(functions):
        print('Ошибка! Указана неправильная функция.')
    else:
        x, y = request_numbers()
        result = functions[action](x, y)
        print(f'Результат выполнения функции: {result:.2f}')
else:
    result = None
    x, y = request_numbers()
    match action:
        case '+':
            result = x + y
        case '-':
            result = x - y
        case '*':
            result = x * y
        case '/':
            if y == 0:
                print('Ошибка! На ноль делить нельзя.')
                exit(1)
            result = x / y
        case _:
            print('Неизвестное действие!')
            exit(1)
    print(f'Результат подсчёта: {result:.2f}')
```
